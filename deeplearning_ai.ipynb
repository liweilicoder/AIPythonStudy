```python
# matplotlib
import matplotlib.pyplot as plt
import numpy as np

# 生成 x 轴数据
x = np.linspace(-1, 1, 500)

# 定义自定义函数
y_parabola = x**2 - 4*x + 3  # 抛物线
y_gaussian = np.exp(x**1)    # 高斯函数

# 绘制图像
plt.figure(figsize=(10, 6))

# 绘制抛物线
plt.plot(x, y_parabola, color='green', linewidth=2, label='抛物线')

# 绘制高斯函数
plt.plot(x, y_gaussian, color='purple', linewidth=2, label='高斯函数')

plt.tight_layout()
plt.show()
```

```python
from helper_functions import get_llm_response
from csv_functions import display_table
from IPython.display import Markdown
import csv

f = open("./csv/journal.csv", "r")
csv_reader = csv.DictReader(f)
itinerary = []
for row in csv_reader:
    itinerary.append(row)
    print(row)

f.close()
print("--------------------------------------------------------")
print(itinerary)

display_table(itinerary)

filtered_data = []
# Filter by country
for trip_stop in itinerary:
    # For example: get the destinations located in "Japan"
    if trip_stop["Country"] == "Japan":
        filtered_data.append(trip_stop)
display_table(filtered_data)

trip_stop = itinerary[0]
city = trip_stop["City"]
country = trip_stop["Country"]
arrival = trip_stop["Arrival"]
departure = trip_stop["Departure"]

prompt = f"""I will visit {city}, {country}, from {arrival} to {departure}.
Please create a detailed daily itinerary."""

# Store the LLM response
response = get_llm_response(prompt, "minimax")

# Print in Markdown format
display(Markdown(response))
```
