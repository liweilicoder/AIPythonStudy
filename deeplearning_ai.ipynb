```python
# matplotlib
import matplotlib.pyplot as plt
import numpy as np

# 生成 x 轴数据
x = np.linspace(-1, 1, 500)

# 定义自定义函数
y_parabola = x**2 - 4*x + 3  # 抛物线
y_gaussian = np.exp(x**1)    # 高斯函数

# 绘制图像
plt.figure(figsize=(10, 6))

# 绘制抛物线
plt.plot(x, y_parabola, color='green', linewidth=2, label='抛物线')

# 绘制高斯函数
plt.plot(x, y_gaussian, color='purple', linewidth=2, label='高斯函数')

plt.tight_layout()
plt.show()
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv('.env', override=True)
api_key = os.getenv('MINIMAX_API_KEY')
base_url = os.getenv('MINIMAX_BASE_URL')
model = os.getenv('MINIMAX_MODEL')

client = OpenAI(api_key = api_key, base_url = base_url)

def get_llm_response(prompt):
    completion = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "system",
                "content": "You are an AI assistant.You speak always end with a emoij.You answer by Chinese",
            },
            {"role": "user", "content": prompt},
        ],
        temperature=0.0,
    )
    response = completion.choices[0].message.content
    return response

prompt = "你是什么模型？"
response = get_llm_response(prompt)
print(response)
```
